# Scrapeando letras de tango (2)

Scrapper para bajar todas las letras de tango de [El Recodo](https://www.el-recodo.com/). Para eso vamos a usar la librer√≠a `BeatifulSoup`.

```python
import urllib.parse
from bs4 import BeautifulSoup as soup
from numpy import random
import multiprocessing as mp
import pandas as pd
import json, codecs, csv
from google import google
import time
```

```python
t0 = time.time()
print('Start Time: ', time.strftime("%a, %d %b %Y %H:%M:%S %Z", time.localtime(t0)))
n = 13383
tangos = []
for t in range(1,n+1):
    l = len(str(t))
    if l == 1:
        url = 'https://www.el-recodo.com/music?id=0000{}&lang=en'.format(t)
    elif l == 2:
        url = 'https://www.el-recodo.com/music?id=000{}&lang=en'.format(t)
    elif l == 3:
        url = 'https://www.el-recodo.com/music?id=00{}&lang=en'.format(t)
    elif l == 4:
        url = 'https://www.el-recodo.com/music?id=0{}&lang=en'.format(t)
    elif l == 5:
        url = 'https://www.el-recodo.com/music?id={}&lang=en'.format(t)
    
    try:
        r = urllib.request.urlopen(url).read()
    except Exception as e:
        print("type error: " + str(e))
    page_soup = soup(r,"html.parser")
    table = page_soup.findAll('a', {'class':'list-group-item list-group-item-action'})
    lyrics = page_soup.findAll('p', {'class':'text-center'})
    lyrics = lyrics[0].text
    d = {}
    d['URL'] = url
    for i in table:
        col = i.contents[1].text
        val = i.contents[2].string.strip()
        d[col] = val
    d['LYRICS'] = lyrics.strip()
    
    tangos.append(d)
    if t % 100 == 0:
        print(t, '\r')
        print(url)
        keys = tangos[0].keys()
        df = pd.DataFrame(tangos)
        df.to_csv('../Data/letras_elrecodo.csv', index=False)
        t1 = time.time()
        print('Elapsed time:', time.strftime("%H:%M:%S", time.localtime(t1-t0)))
print('End Time: ', time.strftime("%a, %d %b %Y %H:%M:%S %Z", time.localtime(time.time())))
```

```python
with open('../Data/El_Recodo_letras.json', 'w') as fout:
    json.dump(tangos, fout)
```
